```python
# Lets add some more columns in dataframe
customer_name= ['Bigyan Aryal','Monica Diaz','Alborz Ihbrahim','Thomas Maier']
customer_name= random.choices(customer_name,k=1000)
df['Customer_name']= customer_name
df.dtypes
df.head(20)
```

```python
## Updation certain rows of dataframe with car and Fahharad with respective price with .loc function

df.loc[2:5,['Bestellung_name', 'Price($)']] = ['Fahrrad',100]
df.loc[10:12,['Bestellung_name','Price($)']] = ['Audi',5000]
df.head(20)
```

```python
# first Total column
df["Total"] = df['Quantity'] * df['Price($)']

# second Taxes Column
df["Total MwSt"] = df["Total"] + df["Total"] * 0.19
df.head(20)
```
