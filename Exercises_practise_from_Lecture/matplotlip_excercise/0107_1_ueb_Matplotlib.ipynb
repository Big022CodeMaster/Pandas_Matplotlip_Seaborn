```python
# Import von pyplot und numpy
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Erstelle einen array von 0 bis 20 in Steps von 0.4 (x)
x=np.arange(0,20,0.4)
x
```

```python
# f(x) = (x^3 + sin(x)) / 1000
y = (x**3 + np.sin(x)) / 1000
y
```

```python
# erstelle eine Linienplot
plt.plot(y);
```

```python
# Stelle den Plot in grün mit gepunkteter Linie
#plt.plot(x,y,color='green',marker ='.')
plt.plot(x,y,':g')
```

```python
# jeder 5. Punkt soll gleichzeitig zu der grün gepunkteten Linie
# mit einem blauen Dreieck dargestellt werden
plt.plot(x,y,':g',x[::5],y[::5],'^b');
plt.title('my_diagram')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```

```python
# Erstellen neuen Datensatz von 0 bis 4 pi (x) mit 1000 Steps
x_neu= np.linspace(0,4*np.pi,1000)
x_neu[:10]
```

```python
# Erstelle in einer for-Schleife Sinus Kurven
# y1 = np.sin(x)
# y2 = np.sin(2*x)
# y3 = np.sin(3*x)
# y4 = np.sin(4*x)


# Setze Achsenbeschriftung und Titel
# Gestalte frei

plt.figure(figsize=(25,6))
for i in range(5):
    y= np.sin(i*x_neu)

    plt.plot(x_neu,y)

plt.title('Sinuskurven')
plt.xticks([0,np.pi,2*np.pi,3*np.pi,4*np.pi])


plt.show()
```

```python
# y1, y2, y3 und y4 werden wir oben berechnet.
# nun soll jedoch x1 = y1 sein und x2 = y3
# x1 soll mit y2 und x2 mit y4 dargestellt werden.
# (also ursprünglich: y1 mit y2 und y3 mit y4 darstellen)
# Versuche einen Scatterplot

# Freie Formatgestaltung!

for i in range(1,4,2):
    x_neu= np.sin(i* x)
    y_neu= np.sin((i+1) *x)
    plt.scatter(x_neu,y_neu)
    
```
