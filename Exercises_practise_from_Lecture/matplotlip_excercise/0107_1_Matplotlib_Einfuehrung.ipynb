Wir werden in diesem Kurs die Systeme von Matplotlib mit Pyplot, Seaborn und Plotly Go und Express kennen lernen. Diese Systeme werden in **High Level** und **Low Level** unterschieden.

**Low Level**

* Die **Low Level** Systeme bilden auch das Fundament zu den **High Level** Systeme. 
* Die **Low Level** Systeme sind schwerer zu bedienen, da diese mehr Code benötigen.
* zu den **Low Level** Systeme gehören: Matplotlib (Objekt: Figuren- oder Axesobjecte) und Plotly (Objekt: Go-Figure).


**High Level**

* Die **High Level** Systeme bauen auf die **Low Level** Systeme auf. 
* Die **High Level** Systeme sind einfacher zu bedienen. Sie laufen auch schneller, da sie im Hintergrund viele Funktionen automatisch ausfüren.
* zu den **High Level** Systeme gehören: Seaborn (Basis: Matplotlib) und Plotly Express. Beide geben Objekte der Low Level Bibliothek zurück.

Versuchen wir hier eine aus:

```python
# Scatter Plot

# Daten kreieren
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))

# Größe und Farbe der Marker identifizieren
sizes = np.random.uniform(15, 150, len(x))
colors = np.random.uniform(15, 150, len(x))

# Scatterplot erstellen
# c ist ein Array das auf eine Farbpalette zugreift
plt.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

# Achsen definieren
plt.xlim = (0, 8)
plt.ylim = (0, 8)

# Titel
plt.title('Scatterplot')

# Anzeige vom Plot
plt.show()
```

Fangen wir erstmal bei **plt.show()** an: Bisher hat es doch auch funktioniert, den Plot uns anzeigen zu lassen. Eigentlich haben wir jedoch uns das Objekt ausgeben lassen. Richtiger ist es jedoch, die Methode **show()** für die Anzeige zu verwenden. Hierdurch werden alle offenen Figuren angezeigt. Dies ist insbesondere wichtig, wenn wir mehrere Figuren zeichnen.

Kommen wir nun zum **c**. Als Kommentar habe ich oben vermerkt, dass **c** auf eine Farbpalette zugreift. Diese Farbpaletten werden bei Matplotlib **cmap** genannt. Die Default cmap ist die **'viridis'**. Eine Übersicht finden Sie hier: https://matplotlib.org/3.5.0/tutorials/colors/colormaps.html#lightness-of-matplotlib-colormaps

```python
# Scatterplot mit cmap = 'PiYG'

# Daten kreieren
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))

# Größe und Farbe der Marker identifizieren
sizes = np.random.uniform(15, 150, len(x))
colors = np.random.uniform(15, 150, len(x))

# Plotten
plt.scatter(x, y, s=sizes, c=colors, cmap = 'PiYG', vmin=0, vmax=100)

# Achsen definieren
plt.xlim = (0, 8)
plt.ylim = (0, 8)

# Titel
plt.title('Scatterplot mit cmap "PiYG"')

# Anzeige vom Plot
plt.show()
```

### Mehrere Plots in einem Objekt

Beginnen wir zuerst zu einem Bespiel mit zwei Plots in einem Objekt:

```python
# plotting two plots in on plot
# create arrays
x_1, y_1 = np.random.normal(0, 1, (2, 2000))
x_2, y_2 = np.random.normal(2, 1, (2, 2000))

# Plots in eine Zelle schreiben
plt.scatter(x_1, y_2, s = 8, alpha = 0.3, color = 'darkblue');    # Farbe benennen
plt.scatter(x_2, y_2, s = 8, alpha = 0.3, color = '#00CED1');     # HexaCode von Farben
```

```python
# Mehrere Plots u.a. mittels for-Schleife möglich

# Größe der Figure festlegen
plt.figure(figsize = (15, 8))

# for-Schleifen in der die Plots erstellt werden
for i, mue in enumerate([0, 1, 2]):
    for j, sigma in enumerate([0.1, 0.2, 0.3]):
        
        x = np.random.normal(mue, sigma, (1000))
        
        # Erstellung eines Histogramm Plots
        plt.hist(x, alpha = 0.3)
        
# Titel
plt.title('Mehrere Plots in einer Figure')

# Anzeigen
plt.show()
```

Wenn wir ein Haus bauen, benötigen wir ein Fundament auf diesem wir ein Haus aufbauen können. Für unsere Grafik benötigen wir auch ein Fundament. Dieses Fundament ist in Matplotlib das **Figuren Objekt**. Mit diesem Figuren Objekt setzen wir den Rahmen, in dem wir die Grafik platzieren wollen.

In dem **Figuren Objekt** patzieren wir:
* die Hintergrundfarbe (1) und
* bestimmen die Größe der Grafik


```python
# Daten
x = [0.2, 0.4, 0.6]
y = [3,6,9]

# Schritt 1: Figurenobjekt erstellen
fig = plt.figure(facecolor='green') 
```

Nachdem wir das **Figuren Objekt** generiert haben, können wir ein **Achsenobjekt** erstellen. Dieses **Achsenobjekt** wird als **AXES** bezeichnet. In dieses **AXES**-Objekt kannst Du u.a.:
* die Achsen beschriften (3) + (4)
* die Achsen skallieren (5) + (6)
* ...   und vieles mehr ...

Haben wir ein Figuren Objekt gesetzt, kann die Formatierung durch die Methode **set()** vorgenommen werden:

```python
# Schritt 2: AXES-Objekt erstellen
ax = fig.add_subplot()

# Werte dem Axes-Objekt zuordnen
ax.set(xlim=[0, 1],
       ylim=[0, 10], 
       title='- Ich bin das Fundament -',
       xlabel='- Ich bin die X-Achse -',
       ylabel='- Ich bin die Y-Achse -')
```

Jetzt müssen wir die eigentliche Grafik erstellen. Das wird **plotten** genannt. Hier geben wir an:
* Datenbasis
* Art des Plots (Default ist der Line Plot)

```python
# Plot erstellen
ax.plot(x, 
        y,
        color="black")  
```

Zum Abschluss benötigst Du noch den Befehl **plt.show()**, damit die Grafik angezeigt wird.

Das **plt.show()** muss insbesondere dann eingefügt werden, wenn der Plot in einem extra Ausgabefenster angezeigt werden soll. 


```python
# Schritt 3: plt.show() um die Grafik anzeigen zu lassen
plt.show()
```

...und das Ganze müssen wir in eine Zelle schreiben:

```python
# Daten
x = [0.2, 0.4, 0.6]
y = [3,6,9]

# Schritt 1: Figurenobjekt erstellen
fig = plt.figure(facecolor='green')   

# Schritt 2: AXES-Objekt erstellen
ax = fig.add_subplot()

# Werte dem Axes-Objekt zuordnen
ax.set(xlim=[0, 1],
       ylim=[0, 10], 
       title='- Ich bin das Fundament -',
       xlabel='- Ich bin die X-Achse -',
       ylabel='- Ich bin die Y-Achse -')

# Plot erstellen
ax.plot(x, y, color="black")   

# Schritt 3: plt.show() um die Grafik anzeigen zu lassen
plt.show()
```

Diese einfachen Plots eignen sich gut in der Kennenlernphase. Dort wollen wir einen Einblick in die Daten erhalten und es ist nicht so notwendig alles gleich von Anfang an **super hübsch** zu machen.

### subplot ohne 's'

Hier wird nur eine einzelne Subplot-Achse an einer angegebenen Rasterposition erstellt.

Schauen wir uns die an einem Beispiel an. Wir haben hier unser **subplot** vor dem plotten eingefügt. Es gibt an durch die an:

<img src = "subplot.png" width = 300>

1. Ziffer: Anzahl Reihen
2. Ziffer: Anzahl Spalten
3. Ziffer: Position im Grid (Beginnt bei 1)

```python
# Beispiel subplot (ohne s)
plt.figure(figsize = (15, 8))

for i, mue in enumerate([0, 1, 2]):
    for j, sigma in enumerate([0.1, 0.2, 0.3]):
        
        x = np.random.normal(mue, sigma, (1000))
        
        # Erstellung des Plots an einem Rasterpunkt
        plt.subplot(3, 3, 3*i+j+1)
        plt.hist(x)
        
```

### Subplots mit s

Bei **subplots** setzen wir dieses ganz zu Beginn. Wir erstellen somit mit subplot unser **Figuren Objekt**. 

Die Positionierung wird dann durch **ax[Zeile, Spalte]** angeben.

```python
# using subplots instead of figure 
fig, ax = plt.subplots(3, 3,figsize = (15, 8))

for i, mue in enumerate([0, 1, 2]):
    for j, sigma in enumerate([0.1, 0.2, 0.3]):
        
        x = np.random.normal(mue, sigma, (1000,))
        
        ax[i,j].hist(x, 30)
        ax[i,j].set_title(f"$\mu$ = {mue}, $\sigma$ = {sigma}")  # change to set_title
        ax[i,j].set_xlim(-1, 3)                                  # change to set_xlim   
        
#fig.tight_layout()   # Arrangiert automatisch den Abstand zwischen den Plots
```

Wir können es noch etwas verschönern, indem wir die Achsen gemeinsam verwenden:

```python
# share x and y
fig, ax = plt.subplots(3, 3,figsize = (15, 8), sharex = True, sharey = True)

for i, mue in enumerate([0, 1, 2]):
    for j, sigma in enumerate([0.1, 0.2, 0.3]):
        
        x = np.random.normal(mue, sigma, (1000,))
        
        ax[i,j].hist(x, 30)
        ax[i,j].set_title(f"$\mu$ = {mue}, $\sigma$ = {sigma}")  # change to set_title
        ax[i,j].set_xlim(-1, 3)                                  # change to set_xlim   
```

```python
# shape of figure is the set first!
plt.figure(figsize = (7,7))
x = np.linspace(0,1,100)

for i, style in [(1, ':'), (2, '--'), (3, '-.')]:
    y = x ** i
    plt.plot(x,y, label = f"power = {i}", linestyle = style, 
             color = 'gray', linewidth = 4);
    
    plt.title('The three gray lines!')
    plt.xlabel('x values')
    plt.ylabel('y values')
    
plt.legend();
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook

# Load a numpy record array from yahoo csv data with fields date, open, close,
# volume, adj_close from the mpl-data/example directory. The record array
# stores the date as an np.datetime64 with a day unit ('D') in the date column.
price_data = (cbook.get_sample_data('goog.npz', np_load=True)['price_data']
              .view(np.recarray))
price_data = price_data[-250:]  # get the most recent 250 trading days

delta1 = np.diff(price_data.adj_close) / price_data.adj_close[:-1]

# Marker size in units of points^2
volume = (15 * price_data.volume[:-2] / price_data.volume[0])**2
close = 0.003 * price_data.close[:-2] / 0.003 * price_data.open[:-2]

fig, ax = plt.subplots()
ax.scatter(delta1[:-1], delta1[1:], c=close, s=volume, alpha=0.5)

ax.set_xlabel(r'$\Delta_i$', fontsize=15)
ax.set_ylabel(r'$\Delta_{i+1}$', fontsize=15)          #r - row String
ax.set_title('Volume and percent change')

ax.grid(True)
fig.tight_layout()  # 

plt.show()
```
