```python
# Wir arbeiten heute mit dem Iris Datensatz
# Dieser ist sauber und muss daher nicht gro√ü Vorbereitet werden
import seaborn as sns
df = sns.load_dataset('iris')
```

```python
# Scatterplot: sepal_length vs petal_length
plt.scatter(df.sepal_length,df.petal_length);
```

```python
# F√ºhre nochmals den Scatterplot durch, jedoch
# unterscheide farblich nach den Spezies
plt.scatter('sepal_length','petal_length', data=df[df['species']== 'setosa'],c='darkblue')
plt.scatter('sepal_length','petal_length', data=df[df.species== 'versicolor'],c='darkgreen')
plt.scatter('sepal_length','petal_length', data=df[df.species== 'virginica'],c='orange')
#plt.legend(['species ' + str(i) for i in range(1, 4)])

plt.legend(df.species.unique())

plt.show()

#sns.scatterplot('sepal_length','petal_length',hue='species',data=df)

#plt.scatter('sepal_length','petal_length', data=df.species.unique()) will not work
```
