Zu den Figuren auf Achsenebene zählen die Module in der  außer der ersten Reihe:

* scatterplot, lineplot
* histplot, kdeplot, ecdfplot, rugplot

Diese Plot zeichnen in die "Axes". Hier ist das **Figuren Objekt von Matplotlib**. Also können wir hier das Schema von oben verwenden. Lediglich der Plot und die Verhübschungen auf der Achsenebene wir in Seaborn gemacht.

Schauen wir uns hierfür ein Beispiel an:

```python
# Wir wollen 3 swarmplots nebeneinander je Species kreieren
fig, ax = plt.subplots(1, 3, figsize = (15, 6), sharey = True)

data = df            # Wir nehmen unsere Pinguin Datenbank
data = data.rename(columns = {'bill_length_mm': 'Schnabellänge', \
                            'body_mass_g': 'Gewicht'})
color = ['#9bb7d4', '#4d7eb3', '#172636']
species_list = df.species.unique()     # Liste der Species

for i in range(len(species_list)):
    
    sns.scatterplot(x = 'Schnabellänge', y = 'Gewicht', \
                    data = data[data.species == species_list[i]], \
                    ax = ax[i], color = color[i])
    ax[i].set_title(species_list[i])
    
fig.tight_layout() 
```

Scheint also nicht so schwer zu sein! Jedoch müssen wir bei den Parametern aufpassen. Hier eine Übersicht (Auszug) vom Scatterplot in Matplotlib und Seaborn:

| Matplotlib | Seaborn | Bedeutung |
| --- | --- | --- | 
| x, y | x, y | Datenposition |
| data | data | Datensatz |
| --- | hue | Farbliche Gruppierung |
| s | --- | Markergröße |
| --- | size | Gruppierung Größe |
| c | --- | Farbe des Markers |
| marker | --- | Markerstyle |

```python
# Einfacher Scatterplot in Matplotlib
plt.scatter(data = df, x='bill_length_mm', 
            y='body_mass_g', 
            alpha = 0.3,  
            s = 30);
```

```python
# Matplotlib: Scatterplot mit farblicher Unterscheidung je Species
color = ['#9bb7d4', '#4d7eb3', '#172636']

plt.scatter(data = df[df.species == 'Adelie'], 
            x='bill_length_mm', y='body_mass_g', \
            s = 20, c = color[0]);

plt.scatter(data = df[df.species == 'Chinstrap'], 
            x='bill_length_mm', y='body_mass_g',\
            s = 20, c = color[1]);

plt.scatter(data = df[df.species == 'Gentoo'], 
            x='bill_length_mm', y='body_mass_g',\
            s = 20, c = color[2]);

plt.show()
```

```python
# Seaborn: farbliche Unterscheidung je Species mit 'hue'
sns.scatterplot(data=df, 
                x='bill_length_mm', y='body_mass_g', 
                hue="species", palette = 'crest');
```

## Funktionen auf Figuren Ebene

Die erste Reihe unseren obigen Bildes der Übersicht, gibt die **Funktionen auf Figuren Ebenen** an:

* relplot (relational)
* displot(distributions)
* catplot(categorical)

Diese Funktionen bilden ein eigenes Grid (FacetGrid - dies kann man auch benutzen alias subplots). Sie verwenden hier nicht mehr Matplotlib, um das Grid zu bauen. Die Figure wird bestimmt durch **height** oder **aspect** (width = aspect * height). 

```python
# Anzeige der Verteilung der Daten
sns.displot(data = df, 
            x='bill_length_mm', 
            y='body_mass_g');
```

```python
# Veränderung der Höhe durch 'height'
sns.displot(data = df, 
            x='bill_length_mm', 
            y='body_mass_g', 
            height = 3);
```

```python
# Veränderung der Breite durch Faktor 'aspect'
sns.displot(data = df, x='bill_length_mm', y='body_mass_g', \
            height = 3, aspect = 2);
```

Die Übersicht oben möchte aussagen, dass alle **Axes Funktionen** über den jeweiligen **Figuren Funktionen** abgerufen werden können. Z.B. können wir mit dem **displot** auch ein **histplot, kdeplot, ecdfplot** oder **rugplot** kreieren. Hierzu müssen wir **kind** hinzufügen:

```python
# Anzeige der Verteilung der Daten
sns.displot(data = df, x='bill_length_mm', y='body_mass_g',
            kind = "kde");
```

Jede **Figuren Funktion** hat natürlich ein Default Wert eingestellt:

* relplot - scatter
* displot - hist
* catplot - strip

Eine nette Eigenschaft ist, dass man durch **col** die Plotte aufteilen kann (vgl. **by** in Matplotlib)

```python
sns.displot(data = df, x='bill_length_mm', y='body_mass_g', \
            kind = "kde", col = 'species');
```

```python
sns.displot(data = df, x='bill_length_mm', y='body_mass_g', \
            kind = "kde", col = 'species', hue = 'island');
```

### Vor- und Nachteile der Figuren Funktionen

| Vorteil | Nachteil |
| --- | --- |
| Einfache Facettierung durch Datenvariablen | Viele Parameter sind nicht in der Funktionssignatur |
| Legende ist außerhalb des Plots by Default | Kann nicht Teil eines größeren Matplotlib Plots sein |
| Leichte Anpassung auf Figuren Level | API unterscheidet sich von Matplotlib |
| Anpassung Parameter der Figurengröße möglich | Anpassung Parameter der Figurengröße möglich |
